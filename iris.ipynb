```python
shopping_df = shopping_df.dropna()
```

```python
shopping_df.count()
```

```python
print(f"Duplicated entries:{shopping_df.duplicated().sum()}")
```

```python
shop_df = shopping_df.drop(['CustomerID'],axis = 1)
shop_df
```

```python
def change_string(member):
    if member =="Yes":
        return 1
    else:
        return 0
    
shop_df['Card Member'] = shop_df['Card Member'].apply(change_string)
shop_df.head()
```

```python
shop_df['Annual Income'] = shop_df['Annual Income']/1000
shop_df.head()
```

```python
shop_df.rename(columns = {'Card Member':'CardMember'},inplace = True)
shop_df.rename(columns = {'Annual Income':'AnnualIncome'},inplace = True)
shop_df.rename(columns = {'Spending Score (1-100)':'SpendingScore'},inplace = True)
```

```python
shop_df
```

```python
output_file = "new_shopping_data.csv"
shop_df.to_csv(output_file, index=False)
```
